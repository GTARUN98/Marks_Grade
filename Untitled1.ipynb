```python
import pandas as pd
```

```python
Row_list=[[2022001,'Abhijeet',65,65,69,81],
          [2022002,'Ajeet',75,75,90,81],
          [2022003,'Amit',75,5,69,87],
          [2022004,'Ranjeet',55,65,79,91],
          [2022005,'Santosh',85,85,60,61],
          [2022006,'Satyam',73,75,68,51],
          [2022007,'Shivam',85,85,50,40],
          [2022009,'Shyam',75,65,69,81],
          [2022010,'Yash',85,75,89,61]
          ]
  
df  = pd.DataFrame(Row_list,columns=['Reg_no','Name','Subject1','Subject2','Subject3','Subject4'])
df['Total'] = ''
df
```

```python
df1 = df.eval('Total = Subject1+Subject2+Subject3+Subject4') 
df1
```

```python
def grade(value):
  if value/4 >= 90:
    return 'A'
  elif value >=80:
    return 'B'
  elif value >=70:
    return 'C'
  elif value >=50:
    return 'D'
  elif value <50:
    return 'E'


df1['Grade'] = df1['Total'].map(grade)
df1
```

```python
new_df = df1[["Reg_no","Name","Grade"]]
new_df["Grade"]
```

```python
A_list =[]

for index in new_df.index:
  if (new_df["Grade"][index] == 'A'):
    A_list.append(new_df['Name'][index])

A_list
```
